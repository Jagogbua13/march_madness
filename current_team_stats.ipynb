```python
urlc = 'http://www.espn.com/mens-college-basketball/statistics/team/_/stat/assists'
defence = pd.read_html(urlc)
defence[0]
```

```python
count2= 41
d_hold = []
for x in range(0,8):
    urle= f'http://www.espn.com/mens-college-basketball/statistics/team/_/stat/assists/sort/avgAssists/count/{count2}'
    test = pd.read_html(urle)
    d_hold.append(test)
   # hold = test[1:0]
    count2 = count2 + 40
```

```python
d_hold
```

```python
defence_one = defence[0].values.tolist()
defence_one
team = []
gp = []
ast = []
apg = []
to = []
topg = []
ast_to = []
for d_one in defence_one:
    print(d_one)
    team.append(d_one[1])
    gp.append(d_one[2])
    ast.append(d_one[3])
    apg.append(d_one[4])
    to.append(d_one[5])
    topg.append(d_one[6])
    ast_to.append(d_one[7])
```

```python
#team = []
#gp = []
#ast = []
#apg = []
#to = []
#topg = []
#ast_to = []


for dfs in d_hold:
    for stat in dfs:
        orange = stat.values.tolist()
        for stats in orange:
            print(stats)
            team.append(stats[1])
            gp.append(stats[2])
            ast.append(stats[3])
            apg.append(stats[4])
            to.append(stats[5])
            topg.append(stats[6])
            ast_to.append(stats[7])
```

```python
defence_dict = {'team':team,'games_played':gp,"assist":ast,'assist_per_game':apg,'turnover':to
                ,'turnover_per_game':topg,'assist/to':ast_to}
defence_df = pd.DataFrame(defence_dict)
defence_df.head()
```
