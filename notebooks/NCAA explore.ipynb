```python
#creating variables to denote the winner 1 = win 0 =loss
df['h_result'] = np.where(df['h_points_game']>df['a_points_game'],1,0)
df['a_result'] = [1 if x == 0 else 0 for x in df['h_result']]
df.head()
```

```python
corr = df3.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

```python
X= df3[['h_field_goals_att','a_field_goals_att','h_assists','a_assists','h_turnovers','a_turnovers']]
y = df3['h_result']
print(X.describe())
y.head()
```

# Train Test split and first Neural nets

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
print(X_train)
```

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScater model and fit it to the training data
X_scaler = StandardScaler().fit(X_train)
# Transform the training and testing data using the X_scaler

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from keras.utils import to_categorical

# One-hot encoding
y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(units=6, activation='relu', input_dim=6))
model.add(Dense(units=2, activation='softmax'))
```

```python
# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Fit the model to the training data
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=100,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
deep_model = Sequential()
deep_model.add(Dense(units=6, activation='relu', input_dim=6))
deep_model.add(Dense(units=6, activation='relu'))
deep_model.add(Dense(units=2, activation='softmax'))
```

```python
deep_model.compile(optimizer='adam',
                   loss='categorical_crossentropy',
                   metrics=['accuracy'])

deep_model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=100,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = deep_model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(f"Deep Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
deep_results = deep_model.predict(X_test_scaled)
result = np.argmax(deep_results,axis=1,out=None)
#result = result.flatten()
deep_dict = {"Actual":y_test,"predicted":result}
deep_df = pd.DataFrame(deep_dict)
deep_df.head(20)
```

# DIFFERIENCIAL MODELS

```python
#testing  NN's with differential variables
df4= df3[['h_result']]
df4['h_field_goal_att_diff']=df3['h_field_goals_att']- df3[ 'a_field_goals_att']
df4['a_field_goal_att_diff']=df3['a_field_goals_att']- df3['h_field_goals_att']

df4['h_assist_diff']= df3['h_assists']- df3['a_assists']
df4['a_assist_diff']= df3['a_assists']- df3['h_assists']
df4['h_turnover_diff']= df3['h_turnovers']- df3['a_turnovers']
df4['a_turnover_diff']= df3['a_turnovers']- df3['h_turnovers']

```

```python
print(df4.keys())
df4.head()
```

```python
df4.count()
```

```python
X2= df4[['h_field_goal_att_diff', 'a_field_goal_att_diff',
       'h_assist_diff', 'a_assist_diff', 'h_turnover_diff', 'a_turnover_diff']]
y2 = df4['h_result']
```

```python
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, random_state=1)
```

```python
# Create a StandardScater model and fit it to the training data
X2_scaler = StandardScaler().fit(X2_train)
# Transform the training and testing data using the X_scaler

X2_train_scaled = X2_scaler.transform(X2_train)
X2_test_scaled = X2_scaler.transform(X2_test)
```

```python
# One-hot encoding
y2_train_categorical = to_categorical(y2_train)
y2_test_categorical = to_categorical(y2_test)
```

```python
model2 = Sequential()
model2.add(Dense(units=6, activation='relu', input_dim=6))
model2.add(Dense(units=2, activation='softmax'))

model2.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

```

```python
model2.fit(
    X2_train_scaled,
    y2_train_categorical,
    epochs=200,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = model2.evaluate(
    X2_test_scaled, y2_test_categorical, verbose=2)
print(f"Deep Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
deep_model2 = Sequential()
deep_model2.add(Dense(units=6, activation='relu', input_dim=6))
deep_model2.add(Dense(units=6, activation='relu'))
deep_model2.add(Dense(units=6, activation='relu'))
deep_model2.add(Dense(units=6, activation='relu'))
deep_model2.add(Dense(units=2, activation='softmax'))

deep_model2.compile(optimizer='adam',
                   loss='categorical_crossentropy',
                   metrics=['accuracy'])

```

```python
deep_model2.fit(
    X2_train_scaled,
    y2_train_categorical,
    epochs=100,
    shuffle=True,
    verbose=2
)
```

```python
model_loss2, model_accuracy2 = deep_model2.evaluate(
    X2_test_scaled, y2_test_categorical, verbose=2)
print(f"Deep Neural Network - Loss: {model_loss2}, Accuracy: {model_accuracy2}")
```
